```python

def preprocess_quesjson(filename):
    imgtoques  = dict()
    questotext = dict()
    file = open(filename,'r')
    json_file = json.load(file)
    for q in json_file['questions']:
        img_id = q['image_id']
        ques_id = q['question_id']
        ques = q['question']
        
        if img_id not in imgtoques:
            imgtoques[img_id] = list()
        imgtoques[img_id].append(ques_id)
        questotext[ques_id] = ques
    return imgtoques,questotext    
        
```

```python
def preprocess_ansjson(filename):
    quesid_to_ans = dict()
    file = open(filename,'r')
    json_file = json.load(file)
    for q in json_file['annotations']:
        ans = q['multiple_choice_answer']
        ques_id = q['question_id']
        quesid_to_ans[ques_id] = ans
    return quesid_to_ans
    
```

```python
def preprocess(path):
    img = image.load_img(path,target_size = (224,224))
    x = image.img_to_array(img)
    x = np.expand_dims(x,axis = 0)
    x = preprocess_input(x)
    return x
```

```python
def pad_ques(dictionary):
    ques_to_seq = dict()
    for key,val in dictionary.items():
        tokens = nltk.word_tokenize(val)
        tokens = [wordtoix[w] for w in tokens if w in wordtoix]
        tokens = pad_sequences([tokens],max_len,padding = 'post')[0]
        ques_to_seq[key] = tokens
    return ques_to_seq
```

```python

def all_ans(dictionary):
    ans  = dict()
    for key,val in quesid_to_ans.items():
        if val not in ans:
            ans[val] = list()
        ans[val].append(key)    
    return ans    
```

```python
f = open('glove.6B.200d.txt',encoding = 'utf-8')
embedding_vector = {}
for line in f:
    token = line.split()
    word , coeff = token[0],token[1:]
    coeff = np.asarray(coeff,dtype = 'float32')
    embedding_vector[word] = coeff
f.close()    
```

```python
def data_generator(img_to_ques,photo,ques_to_seq,hotvec,BS):
    X1, X2, y = list(), list(), list()
    n = 0
    while 1:
        for key,desc in img_to_ques.items():
            for i in range(len(desc)):
                n+=1
                X1.append(photo[key])
                X2.append(ques_to_seq[desc[i]])
                y.append(hotvec[desc[i]])
                if(n == BS):
                    yield ([array(X1),array(X2)],array(y))
                    X1, X2, y = list(), list(), list()
                    n =  0
```

```python
generator = data_generator(imgtoques,train_features,questoseq,quesid_to_hotvec,2)
```
